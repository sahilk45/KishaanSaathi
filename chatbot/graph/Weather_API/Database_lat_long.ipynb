```python
import asyncio
import asyncpg
import json

# Your Database Connection String
DB_DSN = "postgresql://neondb_owner:npg_qrvhmPKE7g8z@ep-lucky-lake-ambn84ss.c-5.us-east-1.aws.neon.tech:5432/neondb?sslmode=require"

async def check_farmer_fields(farmer_id: str):
    """
    Connects to the database and fetches all fields with Lat/Lon.
    """
    try:
        # 1. Connect to the database
        conn = await asyncpg.connect(DB_DSN)
        print(f"Connected to database successfully.")

        # 2. Execute the fetch query (Added center_lat and center_lon)
        query = """
            SELECT id, field_name, city_name, area_hectares, center_lat, center_lon 
            FROM farm_fields 
            WHERE farmer_id = $1
        """
        rows = await conn.fetch(query, farmer_id)

        # 3. Check the results
        if not rows:
            print(f"No fields found for Farmer ID: {farmer_id}")
        else:
            print(f"✅ Success! Found {len(rows)} fields for this farmer:")
            for r in rows:
                print(f" - [{r['id']}] {r['field_name']}")
                print(f"   📍 Location: {r['center_lat']}, {r['center_lon']}")
                print(f"   🏙️ City: {r['city_name']} | 📏 Area: {r['area_hectares']} ha")
                print("-" * 30)

        await conn.close()
        return rows

    except Exception as e:
        print(f"Error connecting to database: {e}")
        return None

# Example Usage (For Notebooks)
test_farmer_id = "c59f6f44-1a98-4eaa-8cf0-3581316a32bb" 

# Just await it directly in your Notebook cell
await check_farmer_fields(test_farmer_id)
```
