```python
import requests
import json

def get_weather_forecast(lat, lon, days=3):
    """
    Fetches weather forecast from Open-Meteo API.
    No API Key is required for non-commercial use.
    """
    base_url = "https://api.open-meteo.com/v1/forecast"
    
    # Define parameters
    params = {
        "latitude": lat,
        "longitude": lon,
        "daily": "temperature_2m_max,precipitation_sum",
        "forecast_days": days,
        "timezone": "auto" # Automatically adjusts to local time
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status() # Check for HTTP errors
        
        data = response.json()
        
        # Printing the results nicely
        print(f"--- Weather Forecast for ({lat}, {lon}) ---")
        daily = data.get("daily", {})
        times = daily.get("time", [])
        temps = daily.get("temperature_2m_max", [])
        rain = daily.get("precipitation_sum", [])

        for i in range(len(times)):
            print(f"Date: {times[i]} | Max Temp: {temps[i]}°C | Rain: {rain[i]}mm")
            
        return data

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

# Test the function with your provided coordinates
if __name__ == "__main__":
    get_weather_forecast(30.73, 76.78, days=3)
```
